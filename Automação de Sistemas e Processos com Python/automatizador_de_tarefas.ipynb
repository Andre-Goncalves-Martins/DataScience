# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Comandos da biblioteca: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python
import pyautogui as pg #OBS: Essa biblioteca n consegue digitar o '?', por isso tem q usar essa de baixo
import pyperclip as pc
from time import sleep

pg.PAUSE = 1 #por causa da velocidade do processamento dos comandos ai agr ele vai esperar um segundo
# Passo 1 - Entrar no nosso sistema(Entrar link:  https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing)
pg.hotkey('ctrl','t')
link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing'
pc.copy(link) # copia o link para a área de transferencia pra poder colar no navegador quando ele abrir a aba
                # Recomendado para link por causa q usar o pg.write vai dar erro por causa dos caracteres especiais
pg.hotkey('ctrl','v')
pg.press('enter')

# Passo 2 - Navegar no sistema (entrar na pasta Exportar)
sleep(4)
pg.doubleClick(x=416, y=290)

# Passo 3 - Baixar o arquivo de vendas
sleep(4)
pg.click(x=416, y=290)
sleep(4)
pg.click(x=1220, y=194)
sleep(4)
pg.click(x=1029, y=623)
sleep(4)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4 - Calcular faturamento e quantidade de produtos vendidos
import pandas as pd


tabela = pd.read_excel(r'D:\Vendas - Dez.xlsx')# r no começo pra garantir q o python n vai ler nenhum 
                                               # comando especial no caminho do arquivo
faturamento = tabela['Valor Final'].sum()
quantidade = tabela['Quantidade'].sum()
display(tabela)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 5 - Enviar email para a diretoria
pg.hotkey('ctrl','t')
link = 'gmail.com'
pc.copy(link)
pg.hotkey('ctrl','v')
pg.press('enter')

# Clicar no botão escrever
sleep(4)
pg.click(x=102, y=202)

# Escrever o remetente do email
sleep(3)
pg.write('andre.martins@discente.univasf.edu.br')
pg.press('tab')
pg.press('tab')

# Escrever assunto
assunto = 'Relatório de Vendas'
pc.copy(assunto) # Usando essa função pra copiar os dados por conta dos caracteres especiais q tem na variável(nesse caso o 'ó')
pg.hotkey('ctrl','v')
pg.press('tab')

# Escrever o nosso e-mail... OBS: o texto vai ser usando como se fosse comentário pra poder ter várias linhas
texto = f'''
Prezado, boa noite

O faturamento foi de:
R${faturamento:,.2f}

A quantidade de produtos foi de:
{quantidade:,}

att,

André, o melhor
'''
pc.copy(texto)
pg.hotkey('ctrl','v')

# Enviar o email
#pg.hotkey('ctrl','enter')
```
